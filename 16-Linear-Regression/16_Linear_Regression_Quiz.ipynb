```
# corr 0.64
X = df['Avg. Area Income']
y = df['Price']
```

```

X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
```

13)
เทรนโมเดลแบบ Simple Linear Regression (สามารถเลือกคู่ได้ หรือ จะลองทุกคู่ก็ได้)

```
# Linear Model
lm = LinearRegression()
lm.fit(X_train, y_train)
# complete
```

```
print('Intercept:',lm.intercept_)
print('Slope:',lm.coef_)
```

ทุกๆหนึ่งหน่วยของ Avg. Area Income ที่เพิ่มขึ้น ทำให้ราคาเพิ่ม 21.20966364 หน่วย

```
# ขั้นตอนการทำนาย
# นำ X_test มาใส่ โดยไม่ให้โมเดลเห็นผลลัพธ์ใน y_test
predicted = lm.predict(X_test)
predicted
```

ลองทดสอบ 70:30, 80:20, 90:10 ratio
Note: ถ้าข้อมูล train มาก ปกติแล้ว จะ error น้อยลง

```
# corr 0.64
X = df['Avg. Area Income']
y = df['Price']
fig = plt.figure(figsize=(20,10))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
lm = LinearRegression(); lm.fit(X_train, y_train); predicted = lm.predict(X_test)
ax1 = plt.subplot(231)
plt.scatter(X_test, y_test, color='blue', label='data')
plt.plot(X_test, predicted, color='red', label='Predicted Regression line')
plt.xlabel('Avg. Area Income'); plt.ylabel('Price'); plt.legend()
ax4 = plt.subplot(234);sns.distplot((y_test-predicted),bins=20)
ax1.set_title('train/test split 70 : 30',fontsize = 20)
print('train/test split 70 : 30')
print('- RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
print('- Intercept:',lm.intercept_)
print('- Slope:',lm.coef_)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
lm = LinearRegression(); lm.fit(X_train, y_train); predicted = lm.predict(X_test)
ax2 = plt.subplot(232)
plt.scatter(X_test, y_test, color='blue', label='data')
plt.plot(X_test, predicted, color='red', label='Predicted Regression line')
plt.xlabel('Avg. Area Income'); plt.ylabel('Price'); plt.legend()
ax5 = plt.subplot(235);sns.distplot((y_test-predicted),bins=20)
ax2.set_title('train/test split 80 : 20',fontsize = 20)
print('\ntrain/test split 80 : 20')
print('- RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
print('- Intercept:',lm.intercept_)
print('- Slope:',lm.coef_)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=100)
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
lm = LinearRegression(); lm.fit(X_train, y_train); predicted = lm.predict(X_test)
ax3 = plt.subplot(233)
plt.scatter(X_test, y_test, color='blue', label='data')
plt.plot(X_test, predicted, color='red', label='Predicted Regression line')
plt.xlabel('Avg. Area Income'); plt.ylabel('Price'); plt.legend()
ax6 = plt.subplot(236);sns.distplot((y_test-predicted),bins=20)
ax3.set_title('train/test split 90 : 10',fontsize = 20)
print('\ntrain/test split 90 : 10')
print('- RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
print('- Intercept:',lm.intercept_)
print('- Slope:',lm.coef_)
```
