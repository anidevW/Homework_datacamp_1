# Homework 1 - หาราคาทอง
ทำการใช้ beautifulsoup ไปอ่านค่าที่ https://www.goldtraders.or.th/ แล้วเอาราคาทองออกมา (print ออกมา หรือ save ลง file)

```
# Importing Modules
import requests
import bs4
```

```
# Requests
response = requests.get('https://www.goldtraders.or.th/')
print(response.text)
```

```
# Make the Soup
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```
# Navigate the Data Structure
data = soup.find(id='DetailPlace_uc_goldprices1_GoldPricesUpdatePanel')
print(data.prettify())
```

```
# Time of Announced
time = data.findAll('table')[0]
print(time.prettify())
```

```
time_announced = time.find('td').text
print(time_announced)
```

```
# Price
price = data.findAll('table')[1]
print(price.prettify())
```

```
price_data = price.findAll('td')
lst = []
for i in price_data:
  lst.append(i.text)
lst
```

```
lst2 = []
for x in lst:
  tmp = x.replace('\r\n','').\
             replace('\xa0','').\
             replace('\n','').\
             replace('\t','')
  tmp = ' '.join(tmp.split())
  lst2.append(tmp)
lst2
```

```
import numpy as np
lst_new = np.array(lst2).reshape(-1,3)
lst_new
```

```
import pandas as pd
df = pd.DataFrame(lst_new, columns=['','','บาทละ(บาท)'])
df
```

```
# Print Data to Console
print(time_announced)
df
```

# Homework 2 - ราคาหุ้น 
ไปที่  https://www.settrade.com/C13_MarketSummary.jsp?detail=SET50
อ่านค่าตารางข้างล่างทั้งหมด เอาข้อมูลมาเก็บไว้ (print ออกมา หรือ save ลง file)

```
# Importing Modules
import requests
import bs4
```

```
# Requests
response = requests.get('https://www.settrade.com/C13_MarketSummary.jsp?detail=SET50')
print(response.text)
```

```
# Make the Soup
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```
# Navigate the Data Structure
data = soup.find(attrs={'class':'table table-info'})
print(data.prettify())
```

```
# Table
table = data.findAll('td')
lst = []
for i in table:
  lst.append(i.text)
lst
```

```
lst2 = []
for x in lst:
  tmp = x.replace('\r\n','').\
             replace('\xa0','').\
             replace('\n','').\
             replace('\t','')
  tmp = ' '.join(tmp.split())
  lst2.append(tmp)
lst2
```

```
import numpy as np
lst_new = np.array(lst2).reshape(-1,8)
lst_new
```

```
import pandas as pd
df = pd.DataFrame(lst_new,
                  columns=['',
                           'ล่าสุด',
                           'เปลี่ยนแปลง',
                           '% เปลี่ยนแปลง',
                           'สูงสุด',
                           'ต่ำสุด',
                           'ปริมาณ (\'000 หุ้น)',
                           'มูลค่า (ล้านบาท)'])
df
```
