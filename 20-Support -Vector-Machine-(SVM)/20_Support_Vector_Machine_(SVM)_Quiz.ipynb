```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Prostate_Cancer.csv')
```

```
# list comprehension
df['result_title'] = ['Malignant' if x=='M' else 'Benign' for x in df['diagnosis_result']]

# list comprehension เพื่อแปลง classification class ที่อยู่ในรูป string ให้เป็นตัวเลข
df['result'] = [1 if x=='M' else 0 for x in df['diagnosis_result']]

# drop id เพราะไม่ได้ใช้
df.drop('id', axis=1, inplace=True)
```
