```
# Default parameters of Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```
predicted = nb.predict(X_test)
predicted
```

```
# Evaluation
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

print('> Confusion Matrix:\n', confusion_matrix(y_test, predicted))
print('> Accuracy =', accuracy_score(y_test, predicted))
print('> F1-Score =', f1_score(y_test, predicted, average='micro'))
print('> Precision =', precision_score(y_test, predicted, average='micro'))
print('> Recall =', recall_score(y_test, predicted, average='micro'))
```

```
# Feature selection
df_new = df.copy()
df_new.drop(['Ba','Ca','K','Mg','Fe','RI'],axis=1,inplace=True)
```

```
# Normalisation
min_max_scaler_16 = MinMaxScaler()
df_new_minmax = min_max_scaler_16.fit_transform(df_new.drop(['Type'],axis=1))
df_new_norm = pd.DataFrame(df_new_minmax, columns=df_new.columns[:-1])

# Standardisation
sc_16 = StandardScaler()
df_new_std = sc_16.fit_transform(df_new.drop(['Type'],axis=1))
df_new_std = pd.DataFrame(df_new_std, columns=df_new.columns[:-1])

# Non-generalisation
df_new_no = df_new.drop(['Type'],axis=1)
```

```
# Train/Test split
y = df_new['Type']

# Normalised
X_train_new_norm, X_test_new_norm, y_train_new_norm, y_test_new_norm = \
    train_test_split(df_new_norm, y, test_size=0.2, random_state=42)

# Standardised
X_train_new_std, X_test_new_std, y_train_new_std, y_test_new_std = \
    train_test_split(df_new_std, y, test_size=0.2, random_state=42)

# No generalised
X_train_new_no, X_test_new_no, y_train_new_no, y_test_new_no = \
    train_test_split(df_new_no, y, test_size=0.2, random_state=42)
```

```
accuracy_lst_norm = []
accuracy_lst_std = []
accuracy_lst_no = []
lst = np.arange(0.01,1.01,0.01)

for i in lst:
    # Normalised
    nb_new_scaled1 = GaussianNB(var_smoothing=i)
    nb_new_scaled1.fit(X_train_new_norm,y_train_new_norm)
    predicted_i = nb_new_scaled1.predict(X_test_new_norm)
    accuracy_lst_norm.append(accuracy_score(y_test_new_norm, predicted_i))

    # Standardised
    nb_new_scaled2 = GaussianNB(var_smoothing=i)
    nb_new_scaled2.fit(X_train_new_std,y_train_new_std)
    predicted_i = nb_new_scaled2.predict(X_test_new_std)
    accuracy_lst_std.append(accuracy_score(y_test_new_std, predicted_i))

    # No generalised
    nb_new_scaled3 = GaussianNB(var_smoothing=i)
    nb_new_scaled3.fit(X_train_new_no,y_train_new_no)
    predicted_i = nb_new_scaled3.predict(X_test_new_no)
    accuracy_lst_no.append(accuracy_score(y_test_new_no, predicted_i))
```

```
plt.figure(figsize=(10,5))

# Normalised
plt.plot(lst,accuracy_lst_norm,color='black',linestyle='-',marker='o',
        markerfacecolor='black', markersize=7, label='Normalised')

# Standardised
plt.plot(lst,accuracy_lst_std,color='blue',linestyle='-',marker='o',
        markerfacecolor='blue', markersize=7, label='Standardised')

# No generalised
plt.plot(lst,accuracy_lst_no,color='red',linestyle='-',marker='o',
        markerfacecolor='red', markersize=7, label='No generalised')

plt.xlabel('var_smoothing')
plt.ylabel('Accuracy')
plt.ylim([0.54,0.72])
plt.minorticks_on()
plt.grid(which='both', axis='x')
plt.grid(which='major', axis='y')
plt.legend()
plt.title('Hyperparameter tuning (Features Selected)', fontsize=20)
```
