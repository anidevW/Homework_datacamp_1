```
#20
import numpy as np
i = 1
lst_input = []
while len(lst_input) < 9:
    number = int(input('กรอกตัวเลขตัวที่ %d ( ช่วง 10 - 100 และไม่ซ้ำกัน) = ' %i))
    if number in lst_input:
        print('ตัวเลขต้องไม่ซ้ำกันกับ {} ใส่ตัวเลขใหม่ อีกครั้ง'.format(lst_input))
    elif number < 10 or number > 100:
        print('กรุณากรอกตัวเลขในช่วง 1 - 100')
    else:
        lst_input.append(number)
        i = i + 1
print(lst_input)
m = np.array(lst_input).reshape(3,3)
print('สมาชิกของของเมตริก คือ')
print(m)
mT = m.T
print('ผลการทราสโพส คือ')
print(mT)
print('ข้อมูลในตำแหน่ง แถวที่ 2 และ คอลัมภ์ที่่ 2 คือ',mT[2, 2])

```
