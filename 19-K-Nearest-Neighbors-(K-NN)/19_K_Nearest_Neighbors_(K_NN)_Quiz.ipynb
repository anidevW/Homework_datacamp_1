14) ทำ Data Cleaning โดยการลบ หรือ fill average ขึ้นอยู่กับความเหมาะสม

ควรใช้ Normalisation
ควรทำ Data Scaling เพราะแต่ละ features มี scales ที่แตกต่างกัน
เช่น Proline มีช่วง [278,1680] จะกลายเป็นมีความสำคัญกว่า Hue ที่มีช่วง [0.48,1.71]
จึงควรทำ Data Scaling เพื่อให้เกิดความเสมอภาค (ไม่ bias) เมื่อโมเดลทำการ learn

จากการทำ distplot ของแต่ละ feature พบว่า data ไม่เป็น normal distribution
เราจึง ควรใช้ Normalisation

(สำหรับ Standardisation มักจะได้ผลดีกว่ากับ data ที่เป็น normal distribution)

source: https://medium.com/towards-artificial-intelligence/how-when-and-why-should-you-normalize-standardize-rescale-your-data-3f083def38ff

15)
เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

```
# ทำ Normalisation
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
arr_minmax = min_max_scaler.fit_transform(df.drop(['Class'], axis=1))

# สร้าง Data Frame
df_minmax = pd.DataFrame(arr_minmax,columns=df.columns[1:])
df_minmax
```

```
test = min_max_scaler.inverse_transform(arr_minmax)
test = pd.DataFrame(test,columns=df.columns[1:])
test
```

```
# ทำ Standardisation
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_std = sc_X.fit_transform(df.drop(['Class'], axis=1))

# สร้าง Data Frame
df_std = pd.DataFrame(X_std,columns=df.columns[1:])
df_std
```

```
# ไม่ทำไรเลย
df.drop(['Class'], axis=1)
```

12)
สร้าง train/test split ของบ้าน สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจชอบ

```
from sklearn.model_selection import train_test_split

# แบบ Normalisation
X_train_minmax, X_test_minmax, y_train_minmax, y_test_minmax \
= train_test_split(df_minmax, df['Class'], test_size=0.33, random_state=42)

# แบบ Standardisation
X_train_std, X_test_std, y_train_std, y_test_std \
= train_test_split(df_std, df['Class'], test_size=0.33, random_state=42)

# ไม่ทำอะไรเลย
X_train_no_trans, X_test_no_trans, y_train_no_trans, y_test_no_trans \
= train_test_split(df.drop(['Class'], axis=1), df['Class'], test_size=0.33, random_state=42)
```

Model Training

1) แบบ Normalisation

```
from sklearn.neighbors import KNeighborsClassifier

knn_minmax = KNeighborsClassifier(n_neighbors=1)
knn_minmax.fit(X_train_minmax, y_train_minmax)
predicted_minmax = knn_minmax.predict(X_test_minmax)
predicted_minmax
```

2) แบบ Standardisation

```
knn_std = KNeighborsClassifier(n_neighbors=1)
knn_std.fit(X_train_std, y_train_std)
predicted_std = knn_std.predict(X_test_std)
predicted_std
```

3) แบบไม่ทำอะไรเลย

```
knn_no_trans = KNeighborsClassifier(n_neighbors=1)
knn_no_trans.fit(X_train_no_trans, y_train_no_trans)
predicted_no_trans = knn_no_trans.predict(X_test_no_trans)
predicted_no_trans
```

16)
วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน Accuracy, F1 score, Recall, Precision

```
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

print('### Normalised Dataset : K = 1 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_minmax, predicted_minmax))
print('> Accuracy =', accuracy_score(y_test_minmax, predicted_minmax))
print('> F1-Score =', f1_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Precision =', precision_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Recall =', recall_score(y_test_minmax, predicted_minmax, average='micro'))

print('\n### Standardised Dataset : K = 1 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_std, predicted_std))
print('> Accuracy =', accuracy_score(y_test_std, predicted_std))
print('> F1-Score =', f1_score(y_test_std, predicted_std, average='micro'))
print('> Precision =', precision_score(y_test_std, predicted_std, average='micro'))
print('> Recall =', recall_score(y_test_std, predicted_std, average='micro'))

print('\n### Non-Transformed Dataset : K = 1 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_no_trans, predicted_no_trans))
print('> Accuracy =', accuracy_score(y_test_no_trans, predicted_no_trans))
print('> F1-Score =', f1_score(y_test_no_trans, predicted_no_trans, average='micro'))
print('> Precision =', precision_score(y_test_no_trans, predicted_no_trans, average='micro'))
print('> Recall =', recall_score(y_test_no_trans, predicted_no_trans, average='micro'))
```

17)
หาค่า K ที่ดีที่สุด สำหรับ Dataset นี้

```
# ใช้แบบ Normalised Dataset 

# ลองค่า k 1-50 ทดสอบด้วย accuracy
accuracy_lst_all_features = []

for i in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_minmax, y_train_minmax)
    predicted_i = knn.predict(X_test_minmax)
    accuracy_lst_all_features.append(accuracy_score(y_test_minmax, predicted_i))
    
# เนื่องจากข้อมูลมีขนาดเล็กมาก ทำให้ค่าที่ออกมาซ้ำกันเยอะ
plt.figure(figsize=(10,5))
plt.plot(range(1,51),accuracy_lst_all_features,color='black',linestyle='dashed',marker='o',
        markerfacecolor='blue', markersize=7)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.ylim([0.9,1])
plt.minorticks_on()
plt.grid(which='both', axis='x')
plt.grid(which='major', axis='y')
plt.title('Finding the optimal K (ALL FEATURES)', fontsize=20)
```

ค่า K ที่ดีที่สุด สำหรับ แบบ All Features คือ 3-5, 10, 14-18
ซึ่งทำให้ได้ accuracy 98.31%

```
# ให้ค่า K = 4 เพราะได้ accuracy สูงสุด จากกราฟข้างบน
# model training
knn_minmax = KNeighborsClassifier(n_neighbors=4)
knn_minmax.fit(X_train_minmax, y_train_minmax)
predicted_minmax = knn_minmax.predict(X_test_minmax)

# model evaluation
print('### Normalised Dataset : K = 4 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_minmax, predicted_minmax))
print('> Accuracy =', accuracy_score(y_test_minmax, predicted_minmax))
print('> F1-Score =', f1_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Precision =', precision_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Recall =', recall_score(y_test_minmax, predicted_minmax, average='micro'))
```

เลือกเฉพาะ features ที่สนใจมาเทรนโมเดล และวัดผลเปรียบเทียบกับแบบ all-features

```
select_lst = ['Alcohol','Malic_acid','Ash','Magnesium','Nonflavanoid_phenols',
              'Proanthocyanins','Colour_intensity']

# ใช้ y_train_minmax และ y_test_minmax ได้เลยเพราะมันคือเฉลยที่
# มี index ตรงกับ X_train_minmax และ X_test_minmax ที่เรานำมา drop

# model training (ใช้แบบ normalised data)
knn_some = KNeighborsClassifier(n_neighbors=7)
knn_some.fit(X_train_minmax[select_lst], y_train_minmax)
predicted_some = knn_some.predict(X_test_minmax[select_lst])

# model evaluation (Some features)
print('### Normalised Dataset (Some features) : K = 7 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_minmax, predicted_some))
print('> Accuracy =', accuracy_score(y_test_minmax, predicted_some))
print('> F1-Score =', f1_score(y_test_minmax, predicted_some, average='micro'))
print('> Precision =', precision_score(y_test_minmax, predicted_some, average='micro'))
print('> Recall =', recall_score(y_test_minmax, predicted_some, average='micro'))

# model evaluation (All features)
print('\n ### Normalised Dataset (All features) : K = 4 ###')
print('> Confusion Matrix:\n', confusion_matrix(y_test_minmax, predicted_minmax))
print('> Accuracy =', accuracy_score(y_test_minmax, predicted_minmax))
print('> F1-Score =', f1_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Precision =', precision_score(y_test_minmax, predicted_minmax, average='micro'))
print('> Recall =', recall_score(y_test_minmax, predicted_minmax, average='micro'))
```

```
# ลองค่า k 1-50 ทดสอบด้วย accuracy
accuracy_lst_some_features = []

for i in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_minmax[select_lst], y_train_minmax)
    predicted_i = knn.predict(X_test_minmax[select_lst])
    accuracy_lst_some_features.append(accuracy_score(y_test_minmax, predicted_i))
    
# เนื่องจากข้อมูลมีขนาดเล็กมาก ทำให้ค่าที่ออกมาซ้ำกันเยอะ
plt.figure(figsize=(10,5))
plt.plot(range(1,51),accuracy_lst_some_features,color='black',linestyle='dashed',marker='o',
        markerfacecolor='blue', markersize=7)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.ylim([0.8,1])
plt.minorticks_on()
plt.grid(which='both', axis='x')
plt.grid(which='major', axis='y')
plt.title('Finding the optimal K (SOME FEATURES)', fontsize=20)
```

```
# ตัด outlier กับทุก features
df_new = df.copy()
df_new.drop(['Malic_acid','Proanthocyanins','Alcalinity_of_ash'],axis=1,inplace=True)

# ทำ Normalisation
min_max_scaler = MinMaxScaler()
arr_new = min_max_scaler.fit_transform(df_new.drop(['Class'], axis=1))

# สร้าง Data Frame
df_new_minmax = pd.DataFrame(arr_new,columns=df_new.columns[1:])

# train/test split
X_train_new_minmax, X_test_new_minmax, y_train_new_minmax, y_test_new_minmax \
= train_test_split(df_new_minmax, df['Class'], test_size=0.33, random_state=42)

# ลองค่า k 1-50 ทดสอบด้วย accuracy
accuracy_lst_new = []

for i in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_new_minmax, y_train_new_minmax)
    predicted_i = knn.predict(X_test_new_minmax)
    accuracy_lst_new.append(accuracy_score(y_test_new_minmax, predicted_i))
    
# เนื่องจากข้อมูลมีขนาดเล็กมาก ทำให้ค่าที่ออกมาซ้ำกันเยอะ
plt.figure(figsize=(10,5))
plt.plot(range(1,51),accuracy_lst_new,color='black',linestyle='dashed',marker='o',
        markerfacecolor='blue', markersize=7)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.ylim([0.8,1])
plt.minorticks_on()
plt.grid(which='both', axis='x')
plt.grid(which='major', axis='y')
plt.title('Finding the optimal K (Optimised Dataset)', fontsize=20)
```
