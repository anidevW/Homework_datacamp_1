1) สร้าง DataFrame ที่เก็บยอดขายของบริษัท A,B,C,D ของปี 2014,2015,2016,2017 โดยชื่อบริษัทเป็น index และปีเป็น column - สร้างคำสั่ง numpy สุ่มเลขแบบ float ให้เป็นรายได้



```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
values = np.random.randint(50,501,(4,4))*np.random.random_sample(1)
values = np.around(values,2) #ปัดเศษทศนิยม
rows = ['A','B','C','D']
columns = ['2014','2015','2016','2017']
df=pd.DataFrame(values,rows,columns)
df
```

2) ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Bar plot ให้แกน x เป็น ชื่อบริษัท ยอดขายเฉลี่ยเป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```
sns.barplot(x=df.index,y= df.mean(axis=1))
plt.title('Avg.sales')
plt.ylabel('Avg.sales (x10^6 bath)')
```

```
sns.barplot(y=df.index,x= df.mean(axis=1))
plt.title('Avg.sales')
plt.xlabel('Avg.sales (x10^6 bath)')
```

13) สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ

```
new_df1 = pd.crosstab(df1['Pclass'],df1['Sex'])
plt.figure(figsize=(5,4))
sns.heatmap(new_df1,annot=True)
```

```
new_df1
```

14) สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา

```
new_df2 = df1.pivot_table(index = 'Pclass', columns = 'Sex',values = 'Fare')
new_df2
```

```
plt.figure(figsize=(5,4))
sns.heatmap(new_df2,annot=True)
```

15) สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง

```
plt.figure(figsize=(5,4))
sns.heatmap(new_df2,annot=True,linecolor='white',linewidths=2)
```
